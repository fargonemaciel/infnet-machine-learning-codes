```python
# Imports de Bibliotecas
import pandas as pd

from sklearn import preprocessing, neighbors, model_selection, metrics
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.metrics import accuracy_score
```

```python
link = '/content/drive/MyDrive/Colab Notebooks/Arquives/titanic/train.csv'

df_titanic = pd.read_csv(link)
df_titanic.head(10)
```

```python
df_titanic.columns
```

```python
df_titanic.isnull().sum().sort_values(ascending=False)
```

```python
df_titanic['Cabin'] = df_titanic['Cabin'].fillna('Unknown')
df_titanic['Embarked'] = df_titanic['Embarked'].fillna('Unknown')
df_titanic['Age'] = df_titanic['Age'].fillna(df_titanic['Age'].median())
```

```python
df_titanic.isnull().sum().sort_values(ascending=False)
```

```python
feature_list = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Cabin', 'Embarked']
# As features escolhidas são importantes para uma primeira análise exploratória de dados e estudo de correlação entre elas e entre estas e a variável target.
# Temos informações importantes como a classe do passageiro, o portão de embarque e relações familiares que podem mostrar algum padrão digno de nota.

target = ['Survived']
# E como variável target, temos o binário se o passageiro sobreviveu ao acidente ou não.
```

```python
# Separação dos Dados em Teste e Validação/Teste

xtrn, xtst = model_selection.train_test_split(df_titanic, test_size=0.2, stratify=df_titanic[target])

print(xtrn.shape)
print(xtst.shape)
```

```python
print(xtrn[feature_list])
```

```python
xtst.head()
```

```python
var_cat_encoding = ['Sex', 'Cabin', 'Embarked']
```

```python
encoder = LabelEncoder()

for var in var_cat_encoding:
  xtrn[var] = encoder.fit_transform(xtrn[var])

for var in var_cat_encoding:
  xtst[var] = encoder.fit_transform(xtst[var])
```

```python
xtrn.head()
```

```python
xtst.head()
```

```python
# Transformação/Padronização dos Dados

scaler = preprocessing.StandardScaler()
scaler.fit(xtrn[feature_list])

normed_trn = scaler.transform(xtrn[feature_list])
normed_tst = scaler.transform(xtst[feature_list])
```

```python
# Treinamento utilizando Modelo KNN

knn = neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(normed_trn, xtrn[target])
```

```python
# Aplicando o Modelo de Predição

pred_trn = knn.predict(normed_trn)
pred_tst = knn.predict(normed_tst)

print(pred_trn)
print(pred_tst)
```

```python
# Acurácia do Modelo

acc_trn = metrics.accuracy_score(xtrn[target], pred_trn)
acc_tst = metrics.accuracy_score(xtst[target], pred_tst)

print(acc_trn)
print(acc_tst)

# O modelo está performando melhor apenas na base de treino, mas no teste tem baixa acurácia, podendo indicar um baixo k do modelo ou ainda overfitting
# com os dados de trieno.
```

```python
# Testando vários valores de K e comparando os resultados

n_neighbors = [1, 3, 5, 7, 9, 11, 13, 15]

results = []

for k in n_neighbors:
  knn = neighbors.KNeighborsClassifier(n_neighbors=k)
  knn.fit(normed_trn, xtrn[target])
  pred_trn = knn.predict(normed_trn)
  pred_tst = knn.predict(normed_tst)
  acc_trn = metrics.accuracy_score(xtrn[target], pred_trn)
  acc_tst = metrics.accuracy_score(xtst[target], pred_tst)
  results.append([k, acc_trn, acc_tst])

results = pd.DataFrame(results, columns=['K', 'Acurácia Treino', 'Acurácia Teste'])
results

# Agora a acurácia mostra uma melhora significativa para um K maior
```
