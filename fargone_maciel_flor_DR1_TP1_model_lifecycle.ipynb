#### Exercício 4: Aprendizado Baseado em Instâncias vs Modelos
Explique a diferença entre aprendizado baseado em instâncias e aprendizado baseado em modelos, dando um exemplo de cada um.

**Resposta**: Aprendizado baseado em instâncias e aprendizado baseado em modelos, tem a principal diferença nos modelos utilizados. Quando é um modelo baseado em instância, como o KNN, o aprendizado se dá em sua maioria por similaridade ou proximidade, só fazendo as previsões nos novos dados quando o modelo busca os padrões nos dados brutos fornecidos. Se perdermos os dados de histórico, o modelo não consegue continuar prevendo só com os novos inputs. Já os algorítimos baseados em modelos, o aprendizado ocorre em cima de bases de treinamento e ao aplicar fórmulas, as predições sob novos inputs acontecem independentemente de ainda termos os dados de treinamento. **Exemplos**: Predição do preços de imóveis. Se usarmos o KNN que é não paramêtrico, o modelo será baseando em instância e buscará aprender por proximidade quais os valores para novos imóveis que surgirem na base de dados. Se usarmos para o mesmo objetivo, outro modelo como uma Regressão Linear, o modelo já passará por treinamentos de fato e estará apto a estimar os novos preços baseado em padrões observados por meio dos cálculos do modelo.

#### Exercício 5: Desafios em Machine Learning
Enumere e descreva três desafios comuns enfrentados ao criar modelos de Machine Learning.

**Resposta**: Os três desafios que mais vejo são: quantidade de dados de treinamento insuficientes; atributos irrelevantes; overfitting dos dados de treinamento. Quantidade insuficiente de dados para treinar modelos torna as previsões mais pobres e às vezes imprecisas. O sucesso de um modelo de machine learning depende da qualidade dos dados, especialmente da relevância dos atributos (features), sem as quais os modelos vão assumir informações erradas. E para completar os três exemplos, temos o Oberfitting, que nada mais é do que a generalização exacerbada das previsões, em decorrência de inúmeros fatores, como por exemplo, bases muito pequenas com muito ruído nos dados.
